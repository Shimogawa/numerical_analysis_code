## Problem 1



```python
A = np.array([[0, 0, 0, 1],
              [0.5**3, 0.5**2, 0.5, 1],
              [1, 1, 1, 1], 
              [1.5**3, 1.5**2, 1.5, 1],
              [8, 4, 2, 1],
              [2.5**3, 2.5**2, 2.5, 1]])
b = np.array([0, 0.2, 0.27, 0.3, 0.32, 0.33])
```

```python
x_LS = np.linalg.solve(A.T@A, A.T@b)
x_LS
```

```python
pts_x = np.array([0, 0.5, 1, 1.5, 2, 2.5])
```

```python
def genGraph(x_LS, t=3):
    p = np.linspace(0, t, 100)
    py = x_LS @ np.array([p**3, p**2, p, 1])
    return p, py
```

```python
p, py = genGraph(x_LS)
plt.plot(p, py, label='Polynomial')
plt.plot(pts_x, b, '.', label='Data')

# plt.axis('equal')

plt.legend()
# plt.savefig('p1_1.pdf', dpi=100)
```
